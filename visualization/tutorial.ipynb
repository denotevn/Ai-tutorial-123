```python
import matplotlib.pyplot as plt

x_data = [0, 1, 2, 3, 4]
y_data = [1, 3, 4, 5, 3]

#plot 
plt.plot(x_data, y_data)
plt.grid()
plt.ylabel("y values")
plt.xlabel("x values")

# show 
plt.show()
```

```python
x_data = list(range(-50, 50, 1))
y_data = []
for i in x_data:
    y_data.append(i**2)
plt.plot(x_data, y_data)
plt.grid()
```
